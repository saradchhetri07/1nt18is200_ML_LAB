```python
x_mean = df["Hours"].mean()
y_mean = df["Scores"].mean()

x_mean,y_mean
```

```python
df["x"] = df["Hours"] - x_mean
df["y"] = df["Scores"] - y_mean
df["x*y"] = df["x"] * df["y"]
df["x**2"] = df["x"]**2
df["y**2"] = df["y"]**2
```

```python
df
```

```python
summation_x_y = df["x*y"].sum()
summation_x_squared = df["x**2"].sum()
summation_y_squared = df["y**2"].sum()
summation_x_y,summation_x_squared,summation_y_squared
```

```python
r = summation_x_y / ((summation_x_squared)*(summation_y_squared))**0.5
r
```

```python
def mean(numbers):
    if len(numbers) == 0:
        return none
    else:
        sum_numbers = 0
        for i in numbers:
            sum_numbers+=i
            current_avg = sum_numbers / len(numbers)
        return current_avg
    
def get_std(numbers):
    if len(numbers)==0:
        return none
    else:
        new_mean = mean(numbers)
        std_sum = 0
        for i in numbers:
            std_sum+= (i - new_mean)**2
        return std_sum/len(numbers)

std_deviation_x = get_std(df["x"].tolist())**0.5
std_deviation_x

            
# std_deviation_x = (summation_x_squared/len(df["x"]))**0.5
# std_deviation_y = (summation_y_squared/len(df["y"]))**0.5
# std_deviation_x,std_deviation_y
```

```python
m = r * (std_deviation_y/std_deviation_x)
m
```

```python
c = df["Scores"].mean() - m * df["Hours"].mean()
```

```python
c
```

```python
df["y_prediction"] = m*df["Hours"]+c
df
```

```python
plot1 = plt.scatter(df["Hours"],df["y_prediction"])
plot2 = plt.scatter(df["Hours"],df["Scores"])
plt
```

```python
def getSSR(df_pred):
    result = 0
    for i in range(len(df_pred)):
        result += (df_pred[i] - df_pred.mean())**2
    return result
getSSR(df["y_prediction"])
```

```python
def getSSE(df_actual, df_pred):
    result = 0
    for i in range(len(df_actual)):
        result += (df_actual[i] - df_pred[i])**2
    return result
getSSE(df["Scores"], df["y_prediction"])
```
